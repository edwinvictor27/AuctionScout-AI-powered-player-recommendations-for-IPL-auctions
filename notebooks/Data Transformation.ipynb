```python
import pandas as pd
import numpy as np
```

```python
csv_data = r"D:\Git Repositories\AuctionScout-AI-powered-player-recommendations-for-IPL-auctions\dataset\all_bbl_matches_ball_by_ball.csv"
bbl_data = pd.read_csv(csv_data)
bbl_data.tail()
```

```python
bbl_data.info()
```

```python
bbl_data['date'] = pd.to_datetime(bbl_data['date'], dayfirst='True') #converting the string column to date data type column
bbl_data.info()
```

```python
data_from_2020 = bbl_data[bbl_data['date'] >= '10-12-2020']
data_from_2020
```

```python
batsman_stats = data_from_2020.groupby('batsman').agg({"runs_batsman" : "sum"}).reset_index()
batsman_stats.sort_values(by='runs_batsman', ascending=False)
```

```python
batsman_balls_faced = data_from_2020.groupby('batsman').agg({"ball" : "count"}).reset_index()
batsman_balls_faced.sort_values(by="ball", ascending=False)
```

```python
Batsman_stats = pd.merge(batsman_stats, batsman_balls_faced, how='left', on='batsman', sort=True).reset_index()
Batsman_stats.sort_values(by='runs_batsman', ascending=False)
```

```python
data_from_2020["wicket_kind"].unique()
```

```python
Wicket_rows = data_from_2020[data_from_2020["wicket_kind"].notna()]
Wicket_rows
```

```python
wicket_rows_filtered = Wicket_rows[(Wicket_rows['wicket_kind'] != 'run out') & (Wicket_rows['wicket_kind'] != 'retired hurt')]
wicket_rows_filtered 
```

```python
wicket_stats = wicket_rows_filtered['bowler'].value_counts().reset_index()
```

```python
batsman_out_times = Wicket_rows['batsman'].value_counts().reset_index()
```

```python
wicket_stats.columns = ['batsman', 'wickets_taken']
wicket_stats
```

```python
batsman_out_times.columns = ['batsman', 'wickets_lost_count']
batsman_out_times
```

```python
merged_wickets_stats = pd.merge(batsman_out_times, wicket_stats, on='batsman', how='left').fillna(0)
merged_wickets_stats["wickets_taken"] = merged_wickets_stats["wickets_taken"].astype('int64')
merged_wickets_stats
```

```python
player_stats = pd.merge(Batsman_stats, merged_wickets_stats, on='batsman', how='left').fillna(0)
player_stats.sort_values(by=["wickets_taken"], ascending = False)
```

```python
Total_balls_bowled = data_from_2020['bowler'].value_counts().reset_index()
```

```python
Total_balls_bowled.columns = ['batsman', 'Deliveries bowled']
Total_balls_bowled
```

```python
Runs_conceded = data_from_2020.groupby('bowler').agg({"runs_total": "sum"}).reset_index()
Runs_conceded.columns = ['batsman', "runs_conceded"]
Runs_conceded
```

```python
bowlers_df = pd.merge(Total_balls_bowled, Runs_conceded, on='batsman', how='left')
bowlers_df
```

```python
print(bowlers_df.columns)
print(player_stats.columns)
```
